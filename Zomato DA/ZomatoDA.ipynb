```python
# Step 1 --> Importing all Liberaries 
# numpy - for numerical operations
# pandas - for data manipulation and analysis 
# matplotlib and seaborn for visulization 
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Step 2 --> Creating Dataframe from csv file 
```

```python
df = pd.read_csv("zom.csv")
```

```python
df.head(20)
```

```python
# Step 3 --> Check for missing vales in the data 
```

```python
# --> Changing rating from 4.1/5 to 4.1 which is actually needed 
def change(value):
    value = str(value).split('/')
    rate = value[0]
    return rate

df['rate'] = df['rate'].apply(change)
```

```python
df.head(20)
```

```python
# --> Checking for null values , the info fucntion gives info about the data 
df.info()
```

```python
# Q1. What type of restaurant do the majority of customers order from? 
# The best solution  is to plot bar graph to know which type of restraunt is it a cafe , buffe or anything else
# We will use seabron countplot for the same 
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Restraunt")
```

```python
# Conclusion --> Majority people order from dining restraunt 
```

```python
# --> Q2. How many votes has each type of restaurant received from customers?
# we need a graph where we have the type of restraunt and their votes
```

```python
df.head()
```

```python
groupdata = df.groupby('listed_in(type)')['votes'].sum()
plt.plot(groupdata, c="red", marker="o")
plt.xlabel("Restraunt" , c = "Red")
plt.ylabel("Votes" , c = "Red")
```

```python
# Conclusion - Dining Restraunts got max number of votes 
```

```python
# Q6. Which type of restaurant received more offline orders, so that Zomato can customers with some good offers?
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

plt.figure(figsize=(6,6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="Blues")
```
